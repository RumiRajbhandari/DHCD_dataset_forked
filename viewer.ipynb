```python
import torch
import torch.utils.data as data_utils
from dataloader import DHCDataset
import numpy as np
import matplotlib.pyplot as plt

dataManager = DHCDataset('./dataset/dataset.npz')
```

```python
#Helper function to show images
def show_images_grid(imgs_, num_images=5):
    ncols = int(np.ceil(num_images))
    nrows = int(np.ceil(1))
    _, axes = plt.subplots(ncols, nrows, figsize=(nrows, ncols))
    axes = axes.flatten()

    for ax_i, ax in enumerate(axes):
        if ax_i < num_images:
            ax.imshow(imgs_[ax_i][0].numpy(), cmap='Greys_r', interpolation='nearest')
            ax.set_xticks([])
            ax.set_yticks([])
        else:
            ax.axis('off')
```

```python
row_count = 2
col_count = 5
fig, ax = plt.subplots(nrows=row_count, ncols=col_count)
ax = ax.flatten()
for a, img in zip(ax, dataManager.image_train):
    a.imshow(img)
```
